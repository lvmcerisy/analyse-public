```python
import matplotlib.pyplot as plt

n = 25
# Parameters
years = list(range(n+1))  # 0 to 20
population = 100.0
x = 90.0
y = 10.0
growth_rate = 0.006/8*3  # 0.3%
fraction_x = 0.0

# Lists to store results
percent_y = []
percent_y_test = []

# Simulation over 20 years
for year in years:
    # Record percentage of Y
    percent_y.append((y / population) * 100)
    # Compute new population growth
    new_people = population * growth_rate
    var_y = (y + new_people * (1 - fraction_x)) / (population + new_people) - y / population
    percent_y_test.append(var_y * 100)
    # Allocate new people
    x += new_people * fraction_x
    y += new_people * (1 - fraction_x)
    # Update total population
    population += new_people

# Plot percentage of Y over years
plt.figure()
plt.plot(years, percent_y, marker='+')
plt.title("Proportion de Y dans la population sur 20 ans")
plt.xlabel("Ann√©e")
plt.ylabel("Pourcentage de Y (%)")
plt.xticks(years)
plt.grid(True)
plt.tight_layout()
plt.show()
```
