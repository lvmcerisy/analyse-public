```python
import pandas as pd
import matplotlib.pyplot as plt

####################
####################
## IOM 2016-2025  ##
######   UE   ######
####################
####################


# Read the CSV file
df = pd.read_csv("data/migration_arrivals_deaths.csv")
df_sorted = df.sort_values(by="Année").reset_index(drop=True)

df = df_sorted.copy()

df = df[df['Année'] <= 2023 ]

print(df)
```

```python
# Plot the data
plt.figure(figsize=(10, 6))
#plt.plot(df["Année"], df["Arrivées"], marker='o', label='Arrivées')
plt.plot(df["Année"], df["Dead and Missing"]/df["Arrivées"]*100, marker='+', label='Dead and Missing', color='red')
plt.xlabel("Année")
#plt.ylabel("Nombre de personnes")
plt.title("Arrivées et morts/disparus de migrants par année")
#plt.gca().invert_xaxis()  # To show time from past to present
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
