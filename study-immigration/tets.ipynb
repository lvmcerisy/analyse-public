```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Load the CSV data
df = pd.read_csv('clim-base_quot_vent-77-2025.csv')

# Convert date column to datetime format
df['date'] = pd.to_datetime(df['aaaammjj'], format='%Y%m%d')

# Filter data for the specified date range
start_date = '2025-07-01'
end_date = '2025-07-20'
filtered_df = df[(df['date'] >= start_date) & (df['date'] <= end_date)]

# Check if we have any data in this range
if filtered_df.empty:
    print(f"No data available between {start_date} and {end_date}")
else:
    print(f"Found {len(filtered_df)} data points in this range")
    print("Sample data:")
    print(filtered_df[['date', 'rr']].head())

    # Create plot
    plt.figure(figsize=(12, 6))
    plt.plot(filtered_df['date'], filtered_df['rr'], 
             marker='o', linestyle='-', color='royalblue', linewidth=1.5)

    # Format plot
    plt.title('Daily Precipitation in Fontainebleau\n(June 1 - July 20, 2025)', fontsize=14)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Precipitation (mm)', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)

    # Format x-axis dates
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
    plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=7))  # Show weekly ticks
    plt.xticks(rotation=45)
    plt.tight_layout()

    # Save and show plot
    #plt.savefig('fontainebleau_precipitation_june_july_2025.png', dpi=300)
    plt.show()
    
```
